```python
#create a dataframe manually


import pandas as pd
data={
    'Name':['Raj','Poonam','Disha','Kareem','Sanvi'],
    'Age':[25,30,20,29,22],
    'Department':['Sales','Engineering','HR','Enginnering','Sales'],
    'Salary':[50000,60000,40000,65000,50000]
}
df=pd.DataFrame(data)
df
```

```python
#Basic operation

print("First 3 rows:")
df.head(3)
```

```python
print("Column names:")
df.columns
```

```python
print("Data types:")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

```python
#Flitring

print("People with Salary>50000:")
df[df['Salary']>50000]
```

```python
#Sorting

print("Sorted by age:")
df.sort_values(by='Age')
```

```python
#Grouping and aggregation

grouped=df.groupby('Department')['Salary'].mean()
print("Average Salary by Department:")
grouped
```

```python
#Adding new columns

df['Salary in Lakhs']=df['Salary']/100000
df
```

```python
#Line chart


import matplotlib.pyplot as plt
years=[2020,2021,2022,2023]
revenue=[120,150,170,200]
plt.plot(years,revenue,marker='o',linestyle='--',color='red')
plt.title("Company Revenue Over Years",color='green')
plt.xlabel("Year",color='blue')
Plt.ylabel("Revenue",color='blue')
plt.xticks(years)
plt.grid(True)
plt.show()
```

```python
#Bar chart

departments=df['Department'].value_counts()
departments.plot(kind='bar',color='blue')
plt.title("Employee Count by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
#Histogram


ages=df['Age']
plt.hist(ages,bins=5,color='green',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#Pie chart

df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("Department Proportion")
plt.ylabel("")
plt.show()
```

```python
#Load simple dataset

import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

```python
print("\n Info about the tips dataset:")
tips.info()

print("\n Description of the tips dataset")
tips.describe(include='all')
```

```python
#pairplot


sns.pairplot(tips,hue="sex")
plt.show()
```

```python
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.countplot(x='day',data=tips,hue='day',legend=False)
plt.title("Count of Records by Day ")
plt.show()
```

```python
sns.countplot(x='day',data=tips,hue='day',legend=False)
plt.title("Count of Records by Day ")
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```
